```python
common_brands = autos.loc[:,'brand'].value_counts().head(20).index
#getting a list of the 20 most frequent car brands
```

```python
mean_price_dict = {}
for brand in common_brands:
    brand_data = autos[autos.loc[:,'brand']==brand]
    mean = brand_data.loc[:,'price'].mean()
    mean_price_dict[brand] = int(mean)
    
mean_price_dict
#getting the mean price for the 20 most common car brands
```

we aggregated across brands to understand mean price. We observed that in the top 6 brands, there's a distinct price gap.

Audi, BMW and Mercedes Benz are more expensive
Ford and Opel are less expensive
Volkswagen is in between

For the top 6 brands, let's use aggregation to understand the average mileage for those cars and if there's any visible link with mean price. 

```python
mean_mileage_dict ={}
for brand in common_brands:
    brand_data = autos[autos.loc[:,'brand']==brand]
    mean = brand_data.loc[:,'odometer_km'].mean()
    mean_mileage_dict[brand] = mean
    

mean_mileage_dict
```

```python
mean_mileage = pd.Series(mean_mileage_dict).sort_values(ascending=False)
mean_prices = pd.Series(mean_price_dict).sort_values(ascending=False)
```
